```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
```

```python
# Configure logging
logging.basicConfig(filename='web_scraping_errors.log', level=logging.ERROR,
                    format='%(asctime)s - %(levelname)s - %(message)s')
```

```python
# Function to scrape data from a page
def scrape_page(url):
    try:
        response = requests.get(url, headers=headers)
        articles = []  # Initialize a list to store articles

        # Check if response status is 200 OK
        if response.status_code == 200:
            try:
                soup = BeautifulSoup(response.text, 'html.parser')
                menu_items = soup.find_all('div', class_='inner-content-list-right')

                for item in menu_items:
                    # Extracting data with checks to avoid NoneType errors
                    second_a = item.find_all('a')[2]  # Selecting the second <a> tag
                    title = second_a.text.strip()  # Extracting title
                    link = second_a['href']  # Extracting link

                    time_element = item.find('time')
                    content_tag = item.find('p')

                    date = time_element.text.strip() if time_element else 'N/A'
                    content = content_tag.text.strip() if content_tag else 'N/A'

                    # Append the article details as a dictionary
                    articles.append({
                        'Titre': title,
                        'Lien': link,
                        'Date': date,
                        'Contenu': content
                    })

            except Exception as e:
                logging.error(f'Error parsing HTML for {url}: {e}')
                return []  # Return an empty list if parsing fails
        else:
            logging.error(f'Failed to retrieve page: {url} with status code: {response.status_code}')
            return []  # Return an empty list if the request fails

        return articles  # Return the list of articles

    except requests.RequestException as e:
        logging.error(f'Request exception for {url}: {e}')
        return []  # Return an empty list if the request fails
```

```python
#to check all the pages are scrapped 
base_url = 'https://realites.com.tn/fr/category/economie/'
for page_number in range(1,300):
    url = f'{base_url}page/{page_number}/'
   # scrape_page(url)
    print(f"\nScraping Page {page_number}")
    num_articles_on_page = scrape_page(url)

```

```python
# Initialize a list to hold all articles
all_articles = []

# Iterate through pages to accumulate the total count of articles
for page_number in range(1, 300):
    url = base_url + f'page/{page_number}/'
    articles_on_page = scrape_page(url)
    if not articles_on_page:
        print("No articles found, stopping pagination")
        break  # Stop if no articles are found
    all_articles.extend(articles_on_page)  # Add articles from this page to the list

# Create a DataFrame from the list of articles
df = pd.DataFrame(all_articles)

# Display the DataFrame
print(df)
```

```python
# Save the DataFrame to a CSV file
df.to_csv('articles.csv', index=False)
```
